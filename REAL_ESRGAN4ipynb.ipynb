```python
import cv2
import os

raw_klasor = '/content/drive/MyDrive/raw-890'
iyilestirilmis_klasor = '/content/drive/MyDrive/iyilestirilmis-890-v2'
os.makedirs(iyilestirilmis_klasor, exist_ok=True)

dosyalar = os.listdir(raw_klasor)
toplam = len(dosyalar)

for i, dosya in enumerate(dosyalar):
    raw_yol = os.path.join(raw_klasor, dosya)
    cikti_yol = os.path.join(iyilestirilmis_klasor, dosya)

    if os.path.exists(cikti_yol):
        continue

    img = cv2.imread(raw_yol, cv2.IMREAD_COLOR)
    if img is None:
        continue

    output, _ = upsampler.enhance(img, outscale=4)
    cv2.imwrite(cikti_yol, output)
    print(f'{i+1}/{toplam} → {dosya}')

print('Tamamlandı!')
```

```python
import numpy as np
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

raw_klasor = '/content/drive/MyDrive/raw-890'
ref_klasor = '/content/drive/MyDrive/reference-890-extracted/reference-890'
iyi_klasor = '/content/drive/MyDrive/iyilestirilmis-890-v2'

psnr_raw_listesi = []
ssim_raw_listesi = []
psnr_iyi_listesi = []
ssim_iyi_listesi = []

for dosya in os.listdir(raw_klasor):
    raw_yol = os.path.join(raw_klasor, dosya)
    ref_yol = os.path.join(ref_klasor, dosya)
    iyi_yol = os.path.join(iyi_klasor, dosya)

    if not os.path.exists(ref_yol) or not os.path.exists(iyi_yol):
        continue

    raw = cv2.imread(raw_yol)
    ref = cv2.imread(ref_yol)
    iyi = cv2.imread(iyi_yol)

    if raw is None or ref is None or iyi is None:
        continue

    hedef = (raw.shape[1], raw.shape[0])
    ref_r = cv2.resize(ref, hedef)
    iyi_r = cv2.resize(iyi, hedef)

    psnr_raw_listesi.append(psnr(ref_r, raw))
    ssim_raw_listesi.append(ssim(ref_r, raw, channel_axis=2))
    psnr_iyi_listesi.append(psnr(ref_r, iyi_r))
    ssim_iyi_listesi.append(ssim(ref_r, iyi_r, channel_axis=2))

print(f'Karşılaştırılan görüntü: {len(psnr_raw_listesi)}')
print()
print('========== SONUÇLAR ==========')
print(f'HAM GÖRÜNTÜ     → PSNR: {np.mean(psnr_raw_listesi):.2f} dB | SSIM: {np.mean(ssim_raw_listesi):.4f}')
print(f'İYİLEŞTİRİLMİŞ  → PSNR: {np.mean(psnr_iyi_listesi):.2f} dB | SSIM: {np.mean(ssim_iyi_listesi):.4f}')
fark_psnr = np.mean(psnr_iyi_listesi) - np.mean(psnr_raw_listesi)
fark_ssim = np.mean(ssim_iyi_listesi) - np.mean(ssim_raw_listesi)
print(f'FARK            → PSNR: {fark_psnr:+.2f} dB | SSIM: {fark_ssim:+.4f}')
```

Genel: 4x daha agresif iyileştirme yapıyor. Göze güzel görünüyor ama UIEB referansının beklediği sualtı renk düzeltmesini yapmadığı için metrikler yine düşüyor.

```python
import cv2
import numpy as np
import os

def uiqm(img):
    img = img.astype(np.float32) / 255.0
    r, g, b = img[:,:,2], img[:,:,1], img[:,:,0]

    rg = r - g
    yb = (r + g) / 2 - b
    uicm = -0.0268 * np.sqrt(np.mean(rg**2) + np.mean(yb**2)) + 0.1586

    def sobel(ch):
        kx = np.array([[-1,0,1],[-2,0,2],[-1,0,1]], np.float32)
        ky = kx.T
        gx = cv2.filter2D(ch, -1, kx)
        gy = cv2.filter2D(ch, -1, ky)
        return np.abs(gx) + np.abs(gy)

    uism = 0.299 * np.mean(sobel(r)) + 0.587 * np.mean(sobel(g)) + 0.114 * np.mean(sobel(b))
    uiconm = np.log(np.std(img) + 1e-5)

    return 0.282 * uicm + 0.287 * uism + 0.433 * uiconm

def uciqe(img):
    # 0-1 aralığına normalize et
    img_norm = img.astype(np.float32) / 255.0
    img_lab = cv2.cvtColor(img_norm, cv2.COLOR_BGR2Lab)

    l = img_lab[:,:,0] / 100.0      # 0-1 arası
    a = img_lab[:,:,1] / 127.0      # normalize
    b = img_lab[:,:,2] / 127.0      # normalize

    chroma = np.sqrt(a**2 + b**2)
    sigma_c = np.std(chroma)

    mu_l = np.mean(l)

    saturation = chroma / (np.mean(chroma) + 1e-5)
    mu_s = np.std(saturation)

    return 0.4680 * sigma_c + 0.2745 * mu_l + 0.2576 * mu_s

# --- Ana kod ---
raw_klasor = '/content/drive/MyDrive/raw-890'
iyi_klasor = '/content/drive/MyDrive/iyilestirilmis-890-v2'

uiqm_raw = []
uiqm_iyi = []
uciqe_raw = []
uciqe_iyi = []

dosyalar = os.listdir(iyi_klasor)
toplam = len(dosyalar)

for i, dosya in enumerate(dosyalar):
    raw_yol = os.path.join(raw_klasor, dosya)
    iyi_yol = os.path.join(iyi_klasor, dosya)

    if not os.path.exists(raw_yol):
        continue

    raw = cv2.imread(raw_yol)
    iyi = cv2.imread(iyi_yol)

    if raw is None or iyi is None:
        continue

    iyi_r = cv2.resize(iyi, (raw.shape[1], raw.shape[0]))

    uiqm_raw.append(uiqm(raw))
    uiqm_iyi.append(uiqm(iyi_r))
    uciqe_raw.append(uciqe(raw))
    uciqe_iyi.append(uciqe(iyi_r))

    if (i+1) % 50 == 0:
        print(f'{i+1}/{toplam} işlendi...')

print()
print('========== SONUÇLAR ==========')
print(f'HAM GÖRÜNTÜ     → UIQM: {np.mean(uiqm_raw):.4f} | UCIQE: {np.mean(uciqe_raw):.4f}')
print(f'İYİLEŞTİRİLMİŞ  → UIQM: {np.mean(uiqm_iyi):.4f} | UCIQE: {np.mean(uciqe_iyi):.4f}')
fark_uiqm = np.mean(uiqm_iyi) - np.mean(uiqm_raw)
fark_uciqe = np.mean(uciqe_iyi) - np.mean(uciqe_raw)
print(f'FARK            → UIQM: {fark_uiqm:+.4f} | UCIQE: {fark_uciqe:+.4f}')
```

UCIQE:Pozitif! Real-ESRGAN renk canlılığını, parlaklığı ve kontrastı biraz artırdı. Görsel kalite iyileşti.
UIQM:Negatif ve zaten çok düşük. İki nedeni var:
1 — UIQM sualtı görüntüleri için tasarlanmış ama renk dengesi bozuk görüntülerde negatif çıkabiliyor.
2 — Real-ESRGAN keskinleştirme yaparken sualtı renk dengesini daha da bozuyor olabilir.
